## Estadística descriptiva
Caracteriza  un fenómeno que se pretende estudiar cuantitativamente <br>
- Ejemplo: Caracterizar gasto por área administrativa en la SCJN para conocer el gasto promedio, gasto per capita, desviación, máximos, mínimos, estacionalidad, etc
- Dicha caracterización se presenta como un resumen de datos que sean fáciles de leer y de entender, por
ejemplo, en tabulados y gráficos

- La estadística descriptiva intenta definir un objeto de estudio y mostrar el conjunto de atributos que se considera importantes para describir el estado de dicho objeto.

La estadística descriptivas tienen como limitante la cobertura: no pueden cubrir la totalidad de los casos
que conforman un objeto de estudio.
- En muchas situaciones se requiere información de grandes grupos de elementos (individuos, empresas, votantes, hogares, etc.), pero sólo podemos acceder a un cierto número de ellos.
-  Solución: Un estudio del universo a través de una muestra, es decir, del estudio de un número reducido pero seleccionado de casos.

- La inferencia estadística consiste en tomar una muestra para tratar de extrapolar las características de las variables de la muestra al conjunto de los casos.

- Proceso de inferencia: Pasar del caso particular (muestra) al universal (conjunto de los casos).
- Una muestra con un grado de error bajo permite hacer una inferencia mas ajustada a las propiedades del conjunto de los casos.

Permite medir los cambios en el tiempo
- En el caso de los elementos cuyos datos varían de una forma regular en el tiempo, es factible pronosticar su evolución inter temporal en base al estudio de dicha regularidad.
- El comportamiento se puede inferir a partir de datos históricos.
- Por tanto se puede pronosticar el comportamiento de los elementos en el futuro sobre la base de un seguimiento de la información presente e histórica.

Permite que podamos saber qué funciona y qué no funciona como política pública (inferencia causal)
