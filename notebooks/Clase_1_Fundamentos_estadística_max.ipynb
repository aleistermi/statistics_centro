<h1><center>¿Qué es la estadística?</center></h1>

La Estadística se define como una disciplina, basada en determinadas metodologías y conceptos, que consiste en producir, analizar, procesar, interpretar y presentar conjuntoÇ de datos.


<h1><center>¿Qué es un dato?</center></h1>

<h3><center>
Es un valor que representa una característica del objeto de estudio.
</center></h3>

<h3><center>
Ejemplos:
</center></h3>

- Juan tiene 39 años. El dato representa la edad de Juan. 

- El prespuesto de la Suprema Corte de Justicia de la Nación en Julio de 2023 fue de 5,630,534,430.00 pesos. El dato representa el gasto en determinado momento de la SCJN.


<h1><center>¿Qué es una variable?</center></h1>

<h3><center>
Una variable es un conjunto de datos que no son constantes (varían).
</center></h3>

Una variable representa una característica que varía a través de los individuos o entidades de análisis. Por ejemplo, la edad en la Ciudad de México. 


<h1><center>¿Qué es un dato cualitativo?</center></h1>

<h3><center>
Cualidad. Por su naturaleza no es numérica. 
</center></h3>

<h3><center>
Ej. Género, etnicidad, grado de escolaridad, etc...
</center></h3>

<h1><center>¿Qué es un dato cuantitativo?</center></h1>

<h3><center>
Cantidad. Por su naturaleza es numérico. 
</center></h3>

<h3><center>
Ej. Edad, ingreso de las familias, gasto público, presupuesto, etc...
</center></h3>

<h1><center>Diseño muestral</center></h1>


<h3><center>Tamaño de la población</center></h3>

Es la cantidad total de personas en el grupo que deseas estudiar. Si estás tomando una muestra aleatoria de personas en EE .UU., entonces el tamaño aproximado de tu población será de 317 millones. De igual forma, si estás realizando una encuesta en tu empresa, el tamaño de la población es la cantidad total de empleados.

<h3><center>Margen de error</center></h3>

Es un porcentaje que te dice en qué medida puedes esperar que los resultados de tu encuesta reflejen la opinión de la población general. Entre más pequeño sea el margen de error, más cerca estarás de tener la respuesta correcta con un determinado nivel de confianza.

<h3><center>Nivel de confianza</center></h3>

Es un porcentaje que revela cuánta confianza puedes tener en que tu población seleccione una respuesta dentro de un rango determinado. Por ejemplo, un nivel de confianza del 95 % significa que puedes tener una seguridad del 95 % de que los resultados oscilarán entre los números x e y.

## Estimar tamaño de muestra

```python
import scipy.stats as st

poblacion = 1000
nivel_de_confianza = .95 # equivlante a una confianza del 95%
error = 0.05 # un error del 5%


def formula_de_muestreo(poblacion, nivel_de_confianza, error):
    pvalues = {.99:2.58, .95: 1.96, .90:1.65}
    z = pvalues[nivel_de_confianza]
    sigma = .5
    n = ((z**2) * sigma**2 * poblacion) / (error**2 *(poblacion -1) + z**2 * sigma**2) 
    return n 

formula_de_muestreo(poblacion, .99, .02)
```
